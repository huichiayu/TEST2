```python
import numpy as np
import networkx as nx
import math
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
city = pd.read_csv('City_coord1.csv')

nLat = city['Lat'][:].to_numpy()
nLong = city['Long'][:].to_numpy()

city

```

```python
city_info = {}

c_list = city['Cities'].to_list()

pos = {}

for i in range(len(c_list)):
    pos[c_list[i]] = (nLong[i], nLat[i])
    
pos

G = nx.Graph()

for i in range(len(c_list)):
    G.add_node(c_list[i])
    
nx.draw(G, pos, with_labels = True)
```

```python
conne0 = pd.read_csv('Conne_Matrix.csv')
conn1 = conne0.drop(conne0.columns[0],axis = 1)
connM = conn1.to_numpy()

plt.imshow(connM)
```

```python



rail = []
rail_dst = []
for i in range(np.shape(connM)[0]):
    for j in range(i+1,np.shape(connM)[1]):
        if connM[i,j] == 1:
            
            rLat1 = nLat[i]/(180.0/math.pi)
            rLon1 = nLong[i]/(180.0/math.pi)
            
            rLat2 = nLat[j]/(180.0/math.pi)
            rLon2 = nLong[j]/(180.0/math.pi)            
        
            distance =  1.609344*3963.0 * math.acos((math.sin(rLat1) * math.sin(rLat2)) + \
                                            math.cos(rLat1) * math.cos(rLat2) * math.cos(rLon2 - rLon1))
            
            rail.append((c_list[i],c_list[j]))
            rail_dst.append(distance)
            
ed_dict = {}
for i in range(len(rail)):
    ed_dict[rail[i]] = rail_dst[i]
    

for e in ed_dict.keys():
    G.add_edge(*e, weight=ed_dict[e])

# ed_dict    
# rail
# labels = nx.get_edge_attributes(G,'weight')
labels = ed_dict
```

```python
path = nx.shortest_path(G,source='New York',target='Los Angeles')
path
```
